# Customer Segmentation for Targeted Marketing using ML Algorithm K-Means clustering

## Import Required Libraries

 ## Feature Selection & Preprocessing

```python
X = dframe[['Annual Income (k$)', 'Spending Score (1-100)']]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Determine Optimal Number of Clusters (K) using 

### Elbow method

```python
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=50)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.grid()
plt.show()

warnings.filterwarnings('ignore')
```

*In the above plot it is clear that the elbow is at 5 number of clusters, therefore its understood the optimal number of cluster is 5.*

### Silhouette Score (Validation)

```python
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(X_scaled)
    silhouette_scores.append(silhouette_score(X_scaled, cluster_labels))

plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--')
plt.title('Silhouette Score for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.grid()
plt.show()

warnings.filterwarnings('ignore')
```

*Highest score at K=5, confirming our choice.*

## Apply K-Means Clustering (K=5)

```python
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

dframe['Cluster'] = clusters
print(dframe['Cluster'].value_counts())

import warnings
warnings.filterwarnings('ignore')
```

###  Visualize Clusters

### 2D Scatter Plot (Income vs. Spending Score)

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(
    dframe['Annual Income (k$)'],
    dframe['Spending Score (1-100)'],
    dframe['Age'],
    c=dframe['Cluster'],
    cmap='viridis',
    s=60)

ax.set_xlabel('Annual Income (k$)')
ax.set_ylabel('Spending Score')
ax.set_zlabel('Age')
plt.title('3D Customer Segmentation')
plt.colorbar(scatter)
plt.show()
```
